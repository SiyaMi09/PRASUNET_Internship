# Load the data 

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'D:/Internship/Prasunet/Task 1/API_SP.POP.TOTL_DS2_en_csv_v2_507434.csv'
population_data = pd.read_csv(file_path, skiprows=4)

population_data.head(), population_data.columns
```

###### Skipped the first 4 rows while reading the CSV file (skiprows=4), which likely contained metadata or headers.

# Data Cleaning and Preparation:
• The dataset is cleaned by dropping rows with missing values, ensuring that only countries with population data for 2022 are included.

• Population values are converted to integers for numerical consistency and plotting.

```python
# Extract data for the year 2022
year = '2022'
population_2022 = population_data[['Country Name', year]].dropna()
```

```python
# Convert population to integers for plotting
population_2022[year] = population_2022[year].astype(int)
```

# Visualization
• The bar chart effectively visualizes the population data using matplotlib. 

• Each bar represents a country, and the height of the bar represents the population size in millions (assuming the data is in millions).

```python
# Select the top 10 countries by population for clear visualization
top_countries = population_2022.nlargest(10, year)

# Plotting the bar chart
plt.figure(figsize=(12, 8))
plt.bar(top_countries['Country Name'], top_countries[year], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population in 2022')
plt.title('Top 10 Countries by Population in 2022')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

### Top 10 Countries by Population in 2022: 
• The bar chart shows the top 10 countries with the highest populations in the year 2022. 

• This visualization helps quickly identify which countries have the largest populations.
